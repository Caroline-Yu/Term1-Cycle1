```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=([0.02,0.06,0.08,0.1])
#print(x)
```

```python
y=([0.35,0.33,0.35,0.35])
#print(y)
```

```python
plt.plot(x,y,"x")
plt.xlabel("amplitude(m)")
plt.ylabel("time_period(s)")
plt.ylim([0, 0.5])
xerror=np.array([0.0002,0.0006,0.0008,0.001])
yerror=np.array([0.035,0.033,0.035,0.035])
plt.errorbar(x,y,xerr=xerror,yerr=yerror,fmt='o',capsize=3)

fit_x,cov_x=np.polyfit(x,y,1,w=1/xerror,cov=True)
px=np.poly1d(fit_x)
plt.plot(x,px(x))
print(cov_x)
print(px)


plt.show()
```

```python
x1=([0.067,0.20,0.27,0.34])
y1=([0.35/1.12,0.33/1.12,0.35/1.12,0.35/1.12])
plt.plot(x1,y1,"x")
plt.xlabel("amplitude(rad)")
plt.ylabel("T/T0(s)")
plt.ylim([0, 0.5])

x1error=np.array([0.0002,0.0006,0.0008,0.001])
y1error=np.array([0.035,0.033,0.035,0.035])
plt.errorbar(x1,y1,xerr=x1error,yerr=y1error,fmt='o',capsize=3)


fit_x1,cov_x1=np.polyfit(x1,y1,2,w=1/xerror,cov=True)
print(fit_x1)
px1=np.poly1d(fit_x1)
plt.plot(x1,px1(x1))
plt.show()

```

```python
y2=([0.0031,0.0031,0.0031])
x2=([0.032,0.032,0.032])
plt.plot(x2,y2,"x")
plt.xlabel("l/g(m^2/s^2)")
plt.ylabel("T^2/4pi^2(s^2)")
plt.ylim([0, 0.1])
y2error=np.array([0.004,0.004,0.004])
x2error=np.array([0.0032,0.0032,0.0032])
plt.errorbar(x2,y2,xerr=x2error,yerr=y2error,fmt='o',capsize=3)
fit_x2,cov_x2=np.polyfit(x2,y2,1,w=1/x2error,cov=True)
px2=np.poly1d(fit_x2)
plt.plot(x2,px2(x2))
print(cov_x2)
plt.show()


```
